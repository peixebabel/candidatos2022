```python
import pandas as pd

df = pd.read_csv('dados/consulta_cand_2022_BRASIL.csv', encoding ='latin1', sep=';')
display(df)

df_bens = pd.read_csv('dados/bem_candidato_2022_BRASIL.csv', encoding ='latin1', sep=';')
df_bens = df_bens.groupby('SQ_CANDIDATO')[['DS_TIPO_BEM_CANDIDATO', 'VR_BEM_CANDIDATO']].agg(lambda x: list(x)).reset_index()
display(df_bens)

df_social = pd.read_csv('dados/rede_social_candidato_2022.csv', encoding ='latin1', sep=';')
df_social = df_social[['SQ_CANDIDATO', 'DS_URL']]
df_social = df_social.groupby('SQ_CANDIDATO')['DS_URL'].apply(list).reset_index(name='DS_URL')
display(df_social)
```

```python
df = df.merge(df_social, left_on=['SQ_CANDIDATO'], right_on=['SQ_CANDIDATO'], how='left')
df = df.merge(df_bens,   left_on=['SQ_CANDIDATO'], right_on=['SQ_CANDIDATO'], how='left')
df['FOTO_URL'] = 'fotos/F' + df.SG_UF.astype(str) + df.SQ_CANDIDATO.astype(str)
df.head()
```

```python
cols = ['DT_ELEICAO', 'TP_ABRANGENCIA', 'SG_UF', 'SG_UE', 'NM_UE', 'DS_CARGO', 'SQ_CANDIDATO', 'NR_CANDIDATO', 
        'NM_CANDIDATO', 'NM_URNA_CANDIDATO', 'NM_SOCIAL_CANDIDATO', 'NM_EMAIL', 'SG_PARTIDO', 'NM_PARTIDO', 
        'NM_COLIGACAO', 'DS_COMPOSICAO_COLIGACAO', 'DT_NASCIMENTO', 'NR_IDADE_DATA_POSSE', 'CD_GENERO', 
        'DS_GENERO', 'CD_GRAU_INSTRUCAO', 'DS_GRAU_INSTRUCAO', 'CD_ESTADO_CIVIL', 'DS_ESTADO_CIVIL', 
        'CD_COR_RACA', 'DS_COR_RACA', 'VR_DESPESA_MAX_CAMPANHA', 'DS_OCUPACAO', 'CD_OCUPACAO',  'ST_REELEICAO', 
        'ST_DECLARAR_BENS', 'DS_URL', 'DS_TIPO_BEM_CANDIDATO', 'VR_BEM_CANDIDATO', 'FOTO_URL']

len(cols)
```

```python
df = df[cols]
df
```

```python
df.to_csv('CANDIDATOS_2022_FILTRADO.csv', index=False)
```
